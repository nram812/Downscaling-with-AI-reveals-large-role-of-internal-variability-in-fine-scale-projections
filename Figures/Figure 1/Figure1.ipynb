# Computing the Climate Change Signal
The code below computes the climate change signal for a variety of different indicators

### Ground Truth CC Signal (RX1Day, TXX)

```python
cc_signal_pr = 100 * (All_merged_CCAM.Rx1day.sel(year = slice(2080,2099)).mean("year") - All_merged_CCAM.Rx1day.sel(year = slice(1986,2005)).mean("year"))/All_merged_CCAM.Rx1day.sel(year = slice(1986,2005)).mean("year")
cc_signal_tasmax = (All_merged_CCAM.TXx.sel(year = slice(2080,2099)).mean("year") - All_merged_CCAM.TXx.sel(year = slice(1986,2005)).mean("year"))#/All_merged_CCAM.seasonal_pr_hist
```

### Ground Truth CC Signal (RX10Y, TX10Y)

```python
cc_signal_pr_v2 = 100 * (All_merged_CCAM.Rx1day.rolling(year =10).max().sel(year = slice(2080,2099)).mean("year") - All_merged_CCAM.Rx1day.rolling(year =10).max().sel(year = slice(1986,2005)).mean("year"))/All_merged_CCAM.Rx1day.rolling(year =10).max().sel(year = slice(1986,2005)).mean("year")
cc_signal_tasmax_v2 = (All_merged_CCAM.TXx.rolling(year =10).max().sel(year = slice(2080,2099)).mean("year") - All_merged_CCAM.TXx.rolling(year =10).max().sel(year = slice(1986,2005)).mean("year"))#/All_merged_CCAM.seasonal_pr_hist
```

### Emulator RX1DAY/TXX

```python
cc_signal_pr_ai = 100 * (All_merged.Rx1day.sel(year = slice(2080,2099)).mean("year") - All_merged.Rx1day.sel(year = slice(1986,2005)).mean("year"))/All_merged.Rx1day.sel(year = slice(1986,2005)).mean("year")
cc_signal_tasmax_ai = (All_merged.TXx.sel(year = slice(2080,2099)).mean("year") - All_merged.TXx.sel(year = slice(1986,2005)).mean("year"))#/All_merged_CCAM.seasonal_pr_hist
```

### Emulator RX10y...

```python
cc_signal_pr_ai_v2 = 100 * (All_merged.Rx1day.rolling(year =10).max().sel(year = slice(2080,2099)).mean("year") - All_merged.Rx1day.rolling(year =10).max().sel(year = slice(1986,2005)).mean("year"))/All_merged.Rx1day.rolling(year =10).max().sel(year = slice(1986,2005)).mean("year")
cc_signal_tasmax_ai_v2 = (All_merged.TXx.rolling(year =10).max().sel(year = slice(2080,2099)).mean("year") - All_merged.TXx.rolling(year =10).max().sel(year = slice(1986,2005)).mean("year"))#/All_merged_CCAM.seasonal_pr_hist
```

### Seasonal sIGNALS

```python
cc_signal_pr_seasonal = 100*(All_merged_CCAM.seasonal_pr_future - All_merged_CCAM.seasonal_pr_hist)/ All_merged_CCAM.seasonal_pr_hist
cc_signal_tasmax_seasonal = (All_merged_CCAM.seasonal_tasmax_future - All_merged_CCAM.seasonal_tasmax_hist)
```

```python
cc_signal_pr_seasonal_ai = 100*(All_merged.seasonal_pr_future - All_merged.seasonal_pr_hist)/ All_merged.seasonal_pr_hist
cc_signal_tasmax_seasonal_ai = (All_merged.seasonal_tasmax_future - All_merged.seasonal_tasmax_hist)
```
